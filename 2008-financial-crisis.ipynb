Case Study: Housing Crisis Impact on Employment in Key Industries

This project analyzes the industry-level impact of the 2008 housing crisis in the U.S. using BLS QCEW data (2004–2012).
It focuses on Construction, Finance, and Real Estate — exploring trends in average weekly wages and employment before, during, and after the crisis.

Tools Used:
- BigQuery (SQL)
- Tableau (Data Visualization)

Key Visuals:
- Percent Change in Employment
- Average Weekly Wages
- Percent Change in Wages

Construction

Financial Activities

Real Estate and Rental & Leasing

Crisis periods are color-coded:

This query uses the BLS QCEW Public Dataset from BigQuery to compute the average employment and weekly wages per year for three industries:

Construction (NAICS 23)
